```python
#CREAR UNA AGENDA
DATOS_AGENDA = {}
print("***********************************")
print("******  AGENDA DE CONTACTOS ******")
print("***********************************")
opcion = int(input("Seleccióne una de las siguientes opciones: \n 1. Añadir un contacto \n 2. Buscar un contacto \n 3. Editar un contacto \n 4. Eliminar un contacto \n 5. Mostrar todos los contactos \n"))
while opcion >= 1 or opcion <= 5:


  if opcion == 1:
   print("Has seleccionado la opción 1. Añadir un contacto")
   nombre = input("Ingrese el nombre del contacto: ")
   telefono = input("Ingrese el teléfono del contacto: ")
   DATOS_AGENDA[nombre] = telefono
   print("Contacto agregado correctamente")
   print(DATOS_AGENDA)
   break

  elif opcion == 2:
   print("Has seleccionado la opción 2. Buscar un contacto")
   nombre = input("Ingrese el nombre del contacto que desea buscar: ")
   if nombre in DATOS_AGENDA:
      print("El teléfono del contacto es: ", DATOS_AGENDA[nombre])
   else:
      print("El contacto no existe en la agenda")
   break
  elif opcion == 3:
   print("Has seleccionado la opción 3. Editar un contacto")
   nombre = input("Ingrese el nombre del contacto que desea editar: ")
   if nombre in DATOS_AGENDA:
      telefono = input("Ingrese el nuevo teléfono del contacto: ")
      DATOS_AGENDA[nombre] = telefono
      print("Contacto editado correctamente")
      break
   else:
      print("El contacto no existe en la agenda")
      break
  elif opcion == 4:
   print("Has seleccionado la opción 4. Eliminar un contacto")

   nombre = input("Ingrese el nombre del contacto que desea eliminar: ")
   if nombre in DATOS_AGENDA:
      del DATOS_AGENDA[nombre]
      print("Contacto eliminado correctamente")
   else:
      print("El contacto no existe en la agenda")
      break
```
